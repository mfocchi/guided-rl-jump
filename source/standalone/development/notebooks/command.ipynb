```python
import torch
```

```python
num_envs = 10
```

```python
env_ids = torch.arange(0,num_envs)
```

```python
pose_command_o = torch.zeros(num_envs, 7)
```

```python
pos_x = (-1,1)
pos_y = (-0.6,0.6)
pos_z = (-0.4,0.4)
```

```python
roll = (0, torch.pi / 12)
pitch = (0, torch.pi / 12)
yaw = (-torch.pi / 2, torch.pi / 2)
```

```python
r = torch.empty(len(env_ids))
pose_command_o[env_ids, 0] = r.uniform_(*pos_x)
pose_command_o[env_ids, 1] = r.uniform_(*pos_y)
pose_command_o[env_ids, 2] = r.uniform_(*pos_z)

# -- orientation
euler_angles = torch.zeros_like(pose_command_o[env_ids, :3])
euler_angles[:, 0].uniform_(*roll)
euler_angles[:, 1].uniform_(*pitch)
euler_angles[:, 2].uniform_(*yaw)

# euler_angles[:, 0] = euler_angles[:, 0]
# euler_angles[:, 1] = euler_angles[:, 1]
```

```python
mask = torch.rand((num_envs,1))
mask
```

```python
roll_idx = torch.where(mask >= 0.5)[0]
roll_idx
```

```python
pitch_idx = torch.where(mask < 0.5)[0]
pitch_idx
```

```python
euler_angles
```

```python
euler_angles[pitch_idx, 0] = 0.0
euler_angles[roll_idx, 1] = 0.0
```

```python
euler_angles
```
