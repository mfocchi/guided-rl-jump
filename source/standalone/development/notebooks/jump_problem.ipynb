```python
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Constants
g = 9.81
c_tg = [0.6, 0.3]  # Fixed target configuration

# Generate grid for c_lo's z-component and cd_lo_x
c_lo_z_values = np.linspace(0.2, 0.4, 50)  # 50 values from 0.2 to 0.4
cd_lo_x = np.linspace(0.1, 5, 200)  # Range of x-component of velocity

# Create a meshgrid
cd_lo_x_grid, c_lo_z_grid = np.meshgrid(cd_lo_x, c_lo_z_values)

# Compute the corresponding cd_lo_z values for each combination
cd_lo_z_grid = ((c_tg[1] - c_lo_z_grid) / (c_tg[0] - 0.1)) * cd_lo_x_grid + (1 / 2) * g * (c_tg[0] - 0.1) * (1 / (cd_lo_x_grid + 1e-10))

# Create the surface plot
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

# Labeling the axes
ax.set_xlabel('$\\dot{c}_{lo,x}$', fontsize=12)
ax.set_ylabel('$c_{lo,z}$', fontsize=12)
ax.set_zlabel('$\\dot{c}_{lo,z}$', fontsize=12)
ax.set_title('Surface plot showing lift-off velocity z-component as a function of lift-off position z')

# Plot the surface
surf = ax.plot_surface(cd_lo_x_grid, c_lo_z_grid, cd_lo_z_grid, cmap='viridis', vmin=0, edgecolor='none')

# axins = inset_axes(
#     ax,
#     width="5%",  # width: 5% of parent_bbox width
#     height="80%",  # height: 50%
#     loc="center left",
#     bbox_to_anchor=(1.2, 0., 1, 1),
#     bbox_transform=ax.transAxes,
#     borderpad=0,
# )
# fig.colorbar(surf, cax=axins, ticks=[1, 2, 3])

# # Add a color bar for reference
# fig.colorbar(surf, shrink=0.5)

plt.show()
```
